```python
# Importar as bibliotecas
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Acessar o Yahoo Finance para pegar os dados necessários - cotação BOVESPA e gerar gráfico
#  utilizando apenas a coluna Ajd Close do db.
cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', start='01-01-2021', end='12-15-2021')
display(cotacao_bovespa)
cotacao_bovespa['Adj Close'].plot()
plt.show()
```

```python
# Carregar arquivo xlsx com os dados das empresas a serem pesquisadas.
tabela_empresas = pd.read_excel('empresas.xlsx')
display(tabela_empresas)

# Loop para gerar os dados e gráficos de cada empresa no arquivo xlsx.
for empresa in tabela_empresas['Empresas']:
    print(empresa)
    cotacao = web.DataReader(f'{empresa}.SA', data_source='yahoo', start='01-01-2021', end='12-15-2021')
    display(cotacao)
    cotacao['Adj Close'].plot()
    plt.show()
```
